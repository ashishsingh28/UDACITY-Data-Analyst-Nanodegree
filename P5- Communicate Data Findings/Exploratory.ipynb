### 7. ProsperScore
<b>A custom risk score was built using historical Prosper data to assess the risk of Prosper borrower listings. The output to Prosper users is a Prosper score which ranges from 1 to 11, with 11 being the best, or lowest risk, score. The worst, or highest risk, score, is a 1.<br>Prosper uses both the custom score and the credit reporting agency score together to assess the borrowerâ€™s level of risk and determine estimated loss rates.
For detailed explanation</b>:https://www.prosper.com/plp/general-prosper_score/

```python
plt.figure(figsize=(7, 4))
base_color = sns.color_palette()[0]
sns.countplot(data=loan_clean,x="ProsperScore",color=base_color);
```

**Note:** There are alot of borrower's with prosperScore less than 6.0. Based on this we can say loan was even provided to people with high risk score.
